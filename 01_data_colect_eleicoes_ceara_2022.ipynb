# 2. Rotina de scraping

```python
import dataprocessing.datacolect as dc
```

### 2.1 Preparao para abertura do *browser* 

```python
url = 'https://dadosabertos.tse.jus.br/dataset/resultados-2022'
```

```python
webdriver = dc.Webdriver(url)
```

```python
browser = webdriver.set_driver(headless= True) # o parmetro headless define se o navegador ficar visvel
# quando definido como True, ele no ser visvel.
# directory  o parmetro que define o diretrio em que o arquivo baixado ser destinado.
```

```python
browser.get(url)
```

O navegador  aberto na pgina desejada, no caso, naquela que contm os dados para *scraping*.

## 2.2 Extrao dos dados da eleio no Cear

```python
scrape = dc.Scrape(browser)
```

A clula abaixo faz com que a automao clique e direcione para a pgina que contm o conjunto de dados do estado de interesse.

```python
ceara_page = scrape.get_element(path= '//a[@title= "CE - Votao por seo eleitoral - 2022"]')
ceara_page.click()
```

E o *link* para *download* do arquivo  obtido, como mostrado abaixo:

```python
ceara_link_download = scrape.get_element(path= '//a[@class= "btn btn-primary resource-url-analytics resource-type-None"]')\
                            .get_attribute('href')
```

```python
print(ceara_link_download)
```

# 3. Gerenciamento do *download* do conjunto de dados

A rotina deve ser encerrada assim que o arquivo for baixado e estiver no diretrio de interesse. Aps encerrada, o navegador da automao deve ser encerrado. Para isso:

```python
import dataprocessing.downloadmanager as dm
```

A classe `Manager`  instanciada e os parmetros `link` e `directory` so passados.

```python
manager = dm.Manager(
    link= ceara_link_download, # link do arquivo que ser baixado
    directory= r'C:\Users\Nityananda\Desktop\Datasets' # diretrio em que o arquivo ser salvo
)
```

A clula abaixo executa o *download* do arquivo a partir do link obtido e instanciado na varivel `ceara_link_download`:

```python
file_name = manager.download()
```

E aguarda-se at o encerramento do *download* e a presena do arquivo no diretrio para que o navegador seja fechado e a rotina finalizada:

```python
manager.download_waits(file= file_name)
```

```python
browser.quit()
```

O arquivo baixado est no formato *zip*, portanto, antes de prosseguir para a anlise desse conjunto de dados  necessrio a sua descompactao.
